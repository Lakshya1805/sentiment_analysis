```python
import re
def html(text):
    pattern = re.compile('<.*?>')
    text1 = re.sub(pattern,'',text)
    return text1
```

```python
def characters(text):
    x = ''
    for i in text:
        if i.isalnum():
            x = x + i
        else:
            x = x + ' '
    return x
```

```python
characters(df['review'][0])
```

```python
df['review'] = df['review'].apply(characters)
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
def remove_stopwords(text):
    x = []
    for i in text.split():
        if i not in stop_words:
            x.append(i)
    y = x[:]
    x.clear()
    return y
```

```python
remove_stopwords(df['review'][0])
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def Stem(text):
    y = []
    for i in text:
        y.append(ps.stem(i))
    z = y[:]
    y.clear()
    return z
        
```

```python
df['review'] = df['review'].apply(Stem)
```

```python
def Join(text):
    return " ".join(text)
```

```python
df['review'] = df['review'].apply(Join)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000)
```

```python
x = cv.fit_transform(df['review']).toarray()
```

```python
x.shape
```

```python
y = df.iloc[:,-1]
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
cf1 = GaussianNB()
cf2 = MultinomialNB()
cf3 = BernoulliNB()
```

```python
cf1.fit(x_train,y_train)
cf2.fit(x_train,y_train)
cf3.fit(x_train,y_train)
```

```python
y_pred1 = cf1.predict(x_test)
y_pred2 = cf2.predict(x_test)
y_pred3 = cf3.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Gaussian",accuracy_score(y_test,y_pred1))
print("Multinomial",accuracy_score(y_test,y_pred2))
print("Bernoulli",accuracy_score(y_test,y_pred3))
```
